```python
# 없는 값 채우기 .값을 np.nan 값으로 바꾸고 다시 0으로 바꾸기
import numpy as np 
data = df.replace(".", np.nan)
data1 = data.fillna(0)
```

```python
#데이터에 null 값이 있는지 확인
null = data1.isnull()
for col in  null.columns :
    null_count = null[col].value_counts()
    try:
        print(col, " : ", null[True])
    except :
        print(col, " : ", 0)
```
