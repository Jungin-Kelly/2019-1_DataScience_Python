# 6. 여러 변수를 더하여 을 통한 새로운 변수 만들기

```python
# 'Recency + Frequency + Monetary' 값으로 'RFM 점수 합계' 만들기
data["RFM 점수합계"]=(data["Recency"]+data["Frequency"]+data["Monetary"])
data
```

```python
# ' 상품 만족도 + 매장 만족도 + 서비스 만족도' 값으로 '전체 만족도' 만들기
data["전체 만족도"]=(data["상품 만족도"]+data["매장 만족도"]+data["서비스 만족도"])
data
```

```python
# '친절성+신속성+책임성+정확성+전문성' 값으로 "직원평가 합계" 항목 추가하기 
data["직원평가 합계"]=(data["친절성"]+data["신속성"]+data["책임성"]+data["정확성"]+data["전문성"])
data
```

```python
# '상품 품질 + 상품 다양성 + 가격 적절성' 값으로 "상품평가 합계" 항목 추가하기 
data["상품평가 합계"]=(data["상품 품질"]+data["상품 다양성"]+data["가격 적절성"])
data
```

```python
# '매장 청결성 + 공간편의성 + 시야 확보성 + 음향 적절성' 값으로 "매장 평가 합계" 항목 추가하기 
data["매장 평가 합계"]=(data["매장 청결성"]+data["공간 편의성"]+data["시야 확보성"]+data["음향 적절성"])
data
```

# 8.  계산을 통해 데이터 간결하게 하기

```python
# '총 매출액' 100만원 단위로 나눠서 표시
data['총 매출액 (만원 단위)'] = (data['총 매출액']//1000000)*100
data['총 매출액 (만원 단위)']
```

# 구간을 나누어 데이터 간결하게 하기

```python
#할인권 사용횟수 5단위로 나눠서 표시하기
data.loc[ df['할인권 사용 횟수'] <  5, '할인권 사용 횟수'] = 0
data.loc[ (df['할인권 사용 횟수'] >=  5) & (df['할인권 사용 횟수'] < 10)] = 5
data.loc[ (df['할인권 사용 횟수'] >=  10) & (df['할인권 사용 횟수'] < 15)] = 10
data.loc[ (df['할인권 사용 횟수'] >=  15) & (df['할인권 사용 횟수'] < 20)] = 15
data.loc[ (df['할인권 사용 횟수'] >=  20) & (df['할인권 사용 횟수'] < 25)] = 20
data.loc[ (df['할인권 사용 횟수'] >=  25) & (df['할인권 사용 횟수'] < 30)] = 25
data.loc[ df['할인권 사용 횟수'] >= 30 ] = 30
data['할인권 사용 횟수']
```

```python
# '방문빈도 ' 10단위로 나눠서 표시하기 
data.loc[ df['방문빈도'] <  10, '방문빈도'] = 0
data.loc[ (df['방문빈도'] >=  10) & (df['방문빈도'] < 20)] =10
data.loc[ (df['방문빈도'] >=  20) & (df['방문빈도'] < 30)] = 20
data.loc[ (df['방문빈도'] >=  30) & (df['방문빈도'] < 40)] = 30
data.loc[ (df['방문빈도'] >=  40) & (df['방문빈도'] < 50)] = 40
data.loc[ (df['방문빈도'] >=  50) & (df['방문빈도'] < 60)] = 50
data.loc[ (df['방문빈도'] >=  60) & (df['방문빈도'] < 70)] = 60
data.loc[ (df['방문빈도'] >=  70) & (df['방문빈도'] < 80)] = 70
data.loc[ (df['방문빈도'] >=  80) & (df['방문빈도'] < 90)] = 80
data.loc[ (df['방문빈도'] >=  90) & (df['방문빈도'] < 100)] = 90
data.loc[ df['방문빈도'] >= 100 ] = 100
data['방문빈도']
```

```python
# '1회 평균매출액' 10000원 미만은 0처리 나머지는 //10000 해서 만원 단위로 표시
data. loc[ df['1회 평균매출액']<10000,'1회 평균매출액'] = 0
data. loc[ df['1회 평균매출액'] >=10000,'1회 평균매출액'] = data['1회 평균매출액']//10000
```

```python
# 값이 변해 있는 것을 확인 할 수 있음.
data['1회 평균매출액'][300:400]
```

```python
# '총 매출액' 100만원 단위로 나눠서 표시
data['순수 매출액'] = (data['순수 매출액']//1000000)*100
```

```python
# 값이 변한 것을 확인 할 수 있음.
data['순수 매출액']
```

```python
# '거래기간' 30으로 나눠서 개월 수로 표시
data['거래기간'] = data['거래기간']//30 
```

```python
# '상품 품질', '상품 다양성 ', '가격 적절성'   시각화해서 가장 부족한 영역 찾기 -> 모든 영역에서 별 차이가 없었지만 상품다양성이 그나마 낮음
from matplotlib import pyplot as plt 
plt.rcParams['font.family'] = 'Malgun Gothic'
x = ['상품 품질', '상품 다양성', '가격 적절성']
y = [data["상품 품질"].mean(), data["상품 다양성"].mean(), data["가격 적절성"].mean()]
plt.title('상품 평가 항목 별 평균',fontname ='Malgun Gothic')
plt.xlabel('평가항목',fontname ='Malgun Gothic')
plt.ylabel('평균 점수', fontname ='Malgun Gothic')
plt.bar(x, y, width = 0.7, color = 'pink')
plt.show()
data["상품 품질"].mean(), data["상품 다양성"].mean(), data["가격 적절성"].mean()
```
